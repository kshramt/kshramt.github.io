# Visualization of large number of data points

```python
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp

fig, ax = plt.subplots()

n = 200000
xs = sp.randn(n)
ys = sp.randn(n)
zs = sp.randn(n)
xs += zs
ys += zs
handle = ax.hist2d(
    xs,
    ys,
    bins=150,
    cmin=1,
)
fig.colorbar(handle[3], label="Frequency")
```
